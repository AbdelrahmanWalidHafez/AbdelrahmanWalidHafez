```python
# data cleaning
newdf = df.dropna()
```

```python
newdf.isnull().sum()
```

```python
newdf=newdf.drop(
    columns=[
    'x_coordinate'
    ,'y_coordinate'
    ,'iucr'
    ,'beat'
    ,'district'
    ,'ward'
    ,'updated_on'
    ,'location'
    ,'year'
    ,'fbi_code'
    ])
```

```python
newdf.info()
```

```python
#transforming data
import holidays
newdf.rename(columns={'unique_key':'id','case_number':'case number','primary_type':'crime type','domestic':'domestic relationship','community_area':'comunit area number'})
newdf['date'] = pd.to_datetime(newdf['date'])
newdf['year'] = newdf['date'].dt.year
newdf['month'] = newdf['date'].dt.month
newdf['day'] = newdf['date'].dt.day

us_holidays = holidays.US()
newdf['holiday'] = newdf['date'].apply(lambda date: date in us_holidays)

newdf['weekend'] = newdf['date'].dt.dayofweek.isin([5, 6])
newdf['normal_date'] = newdf['date'].dt.strftime('%d/%m/%Y')
```

```python
import random
newdf['Gender'] = [random.choice(['Male', 'Female']) for _ in range(len(newdf))]
newdf['Victim Age'] = [random.randint(18, 80) for _ in range(len(newdf))]
```
